```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,100)
```

# Función Escalón
$$
f(x) = (1)*(x > 0)
$$

```python
def escalon(x):
    return 1 * (x > 0)
```

```python
plt.plot(x,escalon(x))
plt.title("FUNCION ESCALON") 
plt.grid()
plt.show()
```

# Función lineal
$$
f(x) = mx+b
$$

```python
def lineal(m,x,b):
    return (m*x) + b
```

```python
plt.plot(x,lineal(2,x, -1))
plt.title("FUNCION LINEAL") 
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.grid()
plt.show()
```

# Función Tangente Hiperbólica
$$
f(x) = \frac{2}{1 + e ^ {-2x}}
$$

```python
def tangenteH(x):
    return 2 / (1 + np.e ** (-2 * x))
```

```python
plt.plot(x,tangenteH(x))
plt.title("FUNCION TANGENTE HIPERBOLICA") 
plt.grid()
plt.show()
```

# Función Sigmoidal
$$
f(x) = \frac{1}{1 + e^{-x}}
$$

```python
def sigmoidal(x):
    return 1/(1 + np.e ** (-x))
```

```python
plt.plot(x,sigmoidal(x))
plt.title("FUNCION SIGMOIDAL") 
plt.grid()
plt.show()
```

# Función Gauss
$$
f(x) = ae ^ - \frac{(x-b)^2}{2c^2}
$$

```python
def gauss(x):
    a = 1
    b = 0
    c = 1
    return a*np.e**-((x -b)**2/2*c**2)
```

```python
plt.plot(x,gauss(x))
plt.title("FUNCION GAUSS") 
plt.grid()
plt.show()
```
