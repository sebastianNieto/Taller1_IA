```python
import matplotlib.pyplot as plt
```

```python
def m(x1, x2, y1, y2):
    return (y2 - y1) / (x2 - x1)
```

# Punto 6

```python
p1_x = [-1,0,1]
p1_y = [1,0,-1]
p2_x = [1,0]
p2_y = [0,1]
x = [-1.5,2]
y = [2,-1.5]
```

```python
plt.plot(p1_x,p1_y, linestyle="none", marker="o", color="red", label="P1")
plt.plot(p2_x,p2_y, linestyle="none", marker="*", color="green", label="P2")
plt.plot(x,y, marker="p")
plt.title(" PUNTO 6")
plt.grid()
plt.show()
```

```python
m(x[0],x[1],y[0],y[1])
```
